```python
# Student Grade Calculator
# This program calculates the average of three subject marks and assigns a grade

print("STUDENT GRADE CALCULATOR")
print("Please enter marks for three subjects (0-100)\n")

# Variable to track if all inputs are valid
valid_input = True

# Get Math marks from user
math_marks = float(input("Enter marks for Math: "))

# Check if Math marks are in valid range (0-100)
if math_marks < 0 or math_marks > 100:
    print("Invalid input. Try again!")
    valid_input = False

# Get Science marks from user (only if previous input was valid)
if valid_input:
    science_marks = float(input("Enter marks for Science: "))
    
    # Check if Science marks are in valid range (0-100)
    if science_marks < 0 or science_marks > 100:
        print("Invalid input. Try again!")
        valid_input = False

# Get English marks from user (only if all previous inputs were valid)
if valid_input:
    english_marks = float(input("Enter marks for English: "))
    
    # Check if English marks are in valid range (0-100)
    if english_marks < 0 or english_marks > 100:
        print("Invalid input. Try again!")
        valid_input = False

# Process and display results only if all inputs are valid
if valid_input:
    # Calculate the average score
    # Formula: Average = (Math + Science + English) / 3
    average = (math_marks + science_marks + english_marks) / 3
    
    # Display the average with 2 decimal places
    print(f"\nYour average score is: {average:.2f}")
    
    # Check for special cases first
    if average == 100:
        # Perfect score - all three subjects must be 100
        print("Wow! You got a perfect score")
    elif average == 0:
        # Zero average - all three subjects must be 0
        print("Oh no! You failed everything")
    # Check grade ranges using if-elif-else statements
    elif average >= 90:
        # Grade A: 90 and above
        print("Excellent! You got an A")
    elif average >= 80:
        # Grade B: 80 to 89
        print("Very Good! You got a B")
    elif average >= 70:
        # Grade C: 70 to 79
        print("Good! You got a C")
    elif average >= 60:
        # Grade D: 60 to 69
        print("Needs Improvement! You got a D")
    else:
        # Grade F: below 60
        print("Fail! You got an F")
```
