```python
# This notebook calculates the area of a circle and triangle based on user input

# Import the math module for pi constant
import math


# CIRCLE AREA CALCULATION
print("CIRCLE AREA CALCULATOR")

# Get radius input from user
# The radius is the distance from the center to the edge of the circle
radius = float(input("Enter the radius of the circle: "))

# Calculate area using formula: Area = π × r²
circle_area = math.pi * radius ** 2

# Display the result with 2 decimal places
print(f"\nThe area of the circle with radius {radius} is: {circle_area:.2f} square units\n")


# TRIANGLE AREA CALCULATION
print("TRIANGLE AREA CALCULATOR")

# Get base and height input from user
# Base is the bottom side of the triangle
# Height is the perpendicular distance from base to opposite vertex
base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))

# Calculate area using formula: Area = (base × height) / 2
triangle_area = (base * height) / 2

# Display the result with 2 decimal places
print(f"\nThe area of the triangle with base {base} and height {height} is: {triangle_area:.2f} square units\n")

# SUMMARY
print("SUMMARY OF CALCULATIONS")
print(f"Circle Area: {circle_area:.2f} square units")
print(f"Triangle Area: {triangle_area:.2f} square units")
```
