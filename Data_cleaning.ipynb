```python

# Imports & basic setup


import pandas as pd
import numpy as np

# Optional: show more columns when displaying DataFrames
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 120)
```

```python

#Load the dataset


file_path = ('C:\\Users\\emili\Downloads\\archive (2)\Superstore.xlsx')   # e.g. "Sample - Superstore.xls" or "data.xlsx"
df = pd.read_excel(file_path)

print("Dataset loaded!")
print("Shape (rows, columns):", df.shape)

df.head()
```

```python

# Quick structure overview


print("=== INFO ===")
df.info()

print("\n=== FIRST 5 ROWS ===")
display(df.head())

print("\n=== LAST 5 ROWS ===")
display(df.tail())
```

```python

# Cell 4 â€“ Check missing values


print("=== Missing values per column ===")
missing_counts = df.isnull().sum()
display(missing_counts)

print("\nColumns with at least 1 missing value:")
display(missing_counts[missing_counts > 0])
```

```python

# Standardize column names
# (lowercase, strip spaces, replace spaces with underscores)

original_columns = df.columns.tolist()
print("Original columns:\n", original_columns)

df.columns = (
    df.columns
      .str.strip()           # remove leading/trailing spaces
      .str.lower()           # lowercase names
      .str.replace(' ', '_') # replace internal spaces with '_'
)

print("\nCleaned columns:\n", df.columns.tolist())
```

```python

# Clean string (object) columns
# (lowercase values, strip spaces)

object_cols = df.select_dtypes(include="object").columns
print("String columns:", list(object_cols))

for col in object_cols:
    df[col] = df[col].str.strip().str.lower()

# Show a small sample to confirm changes
df[object_cols].head()
```

```python

#Convert date columns (if needed)

# Adjust the column names to MATCH your dataset.
# For the Superstore dataset they are usually: 'order_date', 'ship_date'

date_columns = ["order_date", "ship_date"]  # change if your names differ

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")  # invalid dates become NaT

print("Date columns converted (if present).")
df[date_columns].head() if all(c in df.columns for c in date_columns) else print("Some date columns not found.")
```

```python

# Handle missing values (example strategies)


# Check again after conversions
print("Missing values after cleaning:")
display(df.isnull().sum())

# Example of filling numeric missing values with mean
# and categorical/text missing values with "unknown"


num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

# Fill numeric NaNs with column mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical NaNs with "unknown"
df[cat_cols] = df[cat_cols].fillna("unknown")

print("\nMissing values AFTER filling:")
display(df.isnull().sum())
```

```python

#Check and remove duplicates

duplicate_count = df.duplicated().sum()
print("Number of duplicated rows:", duplicate_count)

if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates removed.")
else:
    print("No duplicates found.")

print("New shape (rows, columns):", df.shape)
```

```python

#Basic summary statistics (for verification)


print("=== SUMMARY STATISTICS (NUMERIC) ===")
display(df.describe())

print("\n=== SAMPLE OF CLEANED DATA ===")
display(df.head())
```

```python

 #Save the cleaned dataset


output_csv = "superstore_cleaned.csv"
output_excel = "superstore_cleaned.xlsx"

df.to_csv(output_csv, index=False)
df.to_excel(output_excel, index=False)

print("Cleaned data saved as:")
print("-", output_csv)
print("-", output_excel)
```
